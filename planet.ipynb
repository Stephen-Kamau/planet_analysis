```python
import seaborn as sns
df = sns.load_dataset('planets')
df.head()
df.tail()
```

```python
df.describe()
```

```python
#drop nan values
df.dropna(inplace = False).tail()
```

```python
# We see that most columns were having null values..ie almost half
# lets than check more infor about the data
df.dropna(inplace = False).describe()
```

```python
# from the above we see that most of planet discovery
# was done from year 2009 or later which can be attributed to more advanced tool
# it was started being discovered from year 1989...where     there was slow growth in 
# descovery
df.columns
```

```python
#lets apply groupby functiion using method and median of the orbital_periob
df.groupby('method')['orbital_period'].median()
```

```python
#lets check description of our data by grouping it by years
df.groupby('method')['year'].describe()
```

```python
# from the aove we have seen that the most  planets were discovered through Radial method
# then followed by Transit method
# This maybe attributed to more advanced technology since they are only seen
# to be used in the recent years 
# we can also see that Pulsation Timing variation methods recorded the smallest discoveries
# which may indicate that it was the purest that why it may have been only used for year 2007 only
# and it ceased to be used
```

```python
#lets now count the planets discovered by decades they were discovered
decade = (df['year'] // 10)*10 #divide the years into decades(tens)
decade = decade.astype(str) + 's'#add string "s" to the year
decade.name = 'decade'#add a decade column
#use groupby method to group the planets into the respective decades
df.groupby(['method', decade])['number'].sum().unstack().fillna(0)
#unstact method makes the result into a goodlooking table
```

```python
# from the above we see that most planets are discovered during the decades of 2000s and 2010s
# the discovery of most planets involves the use of Radial velocity and Transit methods
```

```python
#lets the check the sum for each decades by applying sum method
df.groupby(['method', decade])['number'].sum().unstack().fillna(0).sum()
```

```python
#from the above most planets about 60% were discovered in the 2010s decades
#in the 1980s decade only one planet was discovered maybe due to knowledge limitness
#the number of discovery has increased rapidly showing that we will be exxpecting more discoveries in future
#we ahve now seen how discovery startes and grow over time
```
